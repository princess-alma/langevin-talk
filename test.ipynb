```python
from load_mnist import get_mnist_dataset_local
import numpy as np
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = get_mnist_dataset_local()
```

```python
def display_random_image_with_noise(noise_scale=0.1):
    # Pick a random image
    random_index = np.random.randint(0, len(train_images))
    image = train_images[random_index]
    
    # Add noise to the image
    noise = np.random.normal(0, noise_scale, image.shape)
    noisy_image = np.clip(image + noise, 0, 1)  # Ensure pixel values remain in [0, 1]
    
    # Display the original and noisy images side by side
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    axes[1].imshow(noisy_image, cmap='gray')
    axes[1].set_title('Noisy Image')
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
display_random_image_with_noise(noise_scale=1.2)
```
